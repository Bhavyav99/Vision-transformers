```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
from matplotlib import rc
import numpy as np
```

```python
file_path = "metrics.csv"
df = pd.read_csv(file_path)
```

```python
# Train Val Loss Curve

val_loss = []
for i in range(len(df)):
    if df['val_loss'].iloc[i] > 0:
        val_loss.append(df['val_loss'].iloc[i])
        
train_loss = []
for i in range(len(df)):
    if df['val_loss'].iloc[i] > 0:
        train_loss.append(df['train_loss'].iloc[i-1])


plt.figure(figsize=(10, 6))
plt.plot(train_loss, linewidth=2, color='#97C139', label='Train Loss')
plt.plot(val_loss, linewidth=2, color='#004359', label='Validation Loss')
plt.xlabel('Epochs', fontname='Arial', fontsize=14)
plt.ylabel('Loss', fontname='Arial', fontsize=14)
plt.legend()

plt.show()
#plt.savefig("loss.png", dpi=150)
plt.close()
```

```python
# FBeta Score Curve

fb_score_train = []
fb_score_val = []
for i in range(len(df)):
    if df['fbeta_score_train_nok'].iloc[i] > 0:
        fb_score_train.append(df['fbeta_score_train_nok'].iloc[i])
        
for i in range(len(df)):
    if df['fbeta_score_val_nok'].iloc[i] > 0:
        fb_score_val.append(df['fbeta_score_val_nok'].iloc[i])
        
plt.figure(figsize=(10, 6))
plt.plot(fb_score_train, linewidth=2, color='#97C139', label=r'$F_{\beta}$ training score')
plt.plot(fb_score_val, linewidth=2, color='#004359', label=r'$F_{\beta}$ validation score')
plt.xlabel('Epochs', fontname='Arial', fontsize=14)
plt.ylabel('Loss', fontname='Arial', fontsize=14)
plt.legend()
plt.show()
#plt.savefig(r"D:\Bhavya\Plots" + "/" +  "all_parts_train_fbeta_val_fbeta_curve.png", dpi=150)
plt.close()
```

```python
df_cleaned = df.dropna(subset=['precision_train_nok', 'recall_train_nok'])
# Assuming you have columns 'precision_val_nok' and 'recall_val_nok' in the cleaned DataFrame
precision_nok = df_cleaned['precision_train_nok']
recall_nok = df_cleaned['recall_train_nok']

# Plot precision and recall
plt.figure(figsize=(10, 6))
#plt.plot(recall_nok, precision_nok, label='training - Precision-Recall Curve (Class NOK)', linewidth=2)
plt.plot(recall_nok, precision_nok,  linewidth=2, color='#97C139')
#plt.title('Precision-Recall Curve for Class NOK')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.show()
```

```python
##PR Curve for validation dataset


# Filter out rows with NaN values for each epoch
df_cleaned = df.dropna(subset=['precision_val_nok', 'recall_val_nok'])

# Assuming you have columns 'precision_val_nok' and 'recall_val_nok' in the cleaned DataFrame
precision_nok = df_cleaned['precision_val_nok']
recall_nok = df_cleaned['recall_val_nok']

# Plot precision and recall
plt.figure(figsize=(10, 6))
plt.plot(recall_nok, precision_nok, linewidth=2, color= '#004359')
#plt.title('Precision-Recall Curve for Class NOK')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.show()
```
