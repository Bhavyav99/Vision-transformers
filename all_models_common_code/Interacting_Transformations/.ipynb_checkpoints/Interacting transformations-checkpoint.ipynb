```python
import torch
from torchvision import transforms
import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
from IPython.display import display

# Load the image
file_path = "0008@View-001@2020-09-29_10-23-12.jpg.jpg_Part-0002.jpg"  # Replace with your image path
image = cv2.imread(file_path)

# Convert the image to PyTorch tensor
image_tensor = transforms.ToTensor()(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)).unsqueeze(0)

# Function to apply random center crop using PyTorch transforms.RandomCrop
def random_center_crop(image_tensor, crop_size):
    crop_transform = transforms.Compose([
        transforms.RandomCrop(crop_size),
    ])
    cropped_image = crop_transform(image_tensor)
    return cropped_image

# Function to display the cropped image based on the crop size
def display_cropped_image(crop_size):
    cropped_image_tensor = random_center_crop(image_tensor, crop_size)
    cropped_image = cropped_image_tensor.squeeze(0).permute(1, 2, 0).numpy()
    plt.imshow(cropped_image)
    plt.axis('off')
    plt.show()

# Create an interactive slider for adjusting the crop size
interact(display_cropped_image, crop_size=widgets.IntSlider(min=50, max=350, step=10, value=150))


##droppping this transformation 
```

```python
import torch
from torchvision import transforms
import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
from IPython.display import display

# Load the image
file_path = "0008@View-001@2020-09-29_10-23-12.jpg.jpg_Part-0002.jpg"  # Replace with your image path
image = cv2.imread(file_path)

# Convert the image to PyTorch tensor
image_tensor = transforms.ToTensor()(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)).unsqueeze(0)

# Function to apply random grayscale using PyTorch transforms.RandomGrayscale
def random_grayscale(image_tensor, probability):
    grayscale_transform = transforms.Compose([
        transforms.RandomGrayscale(p=probability),
    ])
    grayscaled_image = grayscale_transform(image_tensor)
    return grayscaled_image

# Function to display the grayscaled image based on the probability of grayscale
def display_grayscaled_image(probability):
    grayscaled_image_tensor = random_grayscale(image_tensor, probability)
    grayscaled_image = grayscaled_image_tensor.squeeze(0).permute(1, 2, 0).numpy()
    plt.imshow(grayscaled_image, cmap='gray')  # Display as grayscale
    plt.axis('off')
    plt.show()

# Create an interactive slider for adjusting the probability of grayscale
interact(display_grayscaled_image, probability=widgets.FloatSlider(min=0, max=1, step=0.1, value=0.8))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
from IPython.display import display

# Load the image
file_path = "0008@View-001@2020-09-29_10-23-12.jpg.jpg_Part-0002.jpg"  # Replace with your image path
image = cv2.imread(file_path)

# Function to apply Gaussian blur based on the kernel size
def apply_gaussian_blur(kernel_size):
    blurred_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)
    plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Create an interactive slider for adjusting the kernel size
interact(apply_gaussian_blur, kernel_size=widgets.IntSlider(min=1, max=31, step=2, value=15))
```

```python
import torch
from torchvision import transforms
import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
from IPython.display import display

# Load the image
file_path = "0008@View-001@2020-09-29_10-23-12.jpg.jpg_Part-0002.jpg"  # Replace with your image path
image = cv2.imread(file_path)

# Convert the image to PyTorch tensor
image_tensor = transforms.ToTensor()(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)).unsqueeze(0)

# Function to apply ColorJitter based on the parameters
def apply_color_jitter(brightness, contrast, saturation, hue):
    color_jitter = transforms.ColorJitter(
        brightness=brightness,
        contrast=contrast,
        saturation=saturation,
        hue=hue
    )
    jittered_image = color_jitter(transforms.ToPILImage()(image_tensor.squeeze(0)))
    plt.imshow(jittered_image)
    plt.axis('off')
    plt.show()

# Create interactive sliders for adjusting ColorJitter parameters
interact(
    apply_color_jitter,
    brightness=widgets.FloatSlider(min=0, max=2, step=0.1, value=0.3),
    contrast=widgets.FloatSlider(min=0, max=2, step=0.1, value=0.5),
    saturation=widgets.FloatSlider(min=0, max=2, step=0.1, value=0.2),
    hue=widgets.FloatSlider(min=0.0, max=0.5, step=0.1, value=0.2)
)
```
